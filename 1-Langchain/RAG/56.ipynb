# ollama run gemma:2b

Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = (
    OllamaEmbeddings(model="gemma:2b") # by default it uses llama2
)
embeddings
```

```python
# for running the code, you need to have ollama installed and running with the model "gemma:2b" or "llama2" downloaded.
# Run "ollama run gemma:2b" or "ollama run llama2" in your terminal to start the model server.
r1 = embeddings.embed_documents(
            [
                "Ka is the first letter of the hindi alphabet", 
                "Kha is the second letter of the hindi alphabet"
            ])
print(r1)
```

```python
print(len(r1[0]), " -  ", len(r1[1]))
```

```python
r2 = embeddings.embed_query("What is the first letter of the hindi alphabet?")
print(r2)
```

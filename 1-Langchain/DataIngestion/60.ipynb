```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI() # model="gpt-4o", temperature=0.9, max_tokens=2048)
print(llm)
```

```python
result = llm.invoke("What is generative AI?")
```

```python
print(result)
```

```python
# prompt template -> what role a model wants to play
## chat prompt template -> what role a model wants to play + conversation history

from langchain_core.prompts import ChatPromptTemplate
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert ai engineer. Provide me answers based on the question."),
        ("human", "{input}"),
    ]
)
prompt
```

```python
# create a chain
chain = prompt | llm
response = chain.invoke({"input": "Can you tell me about langsmith?"})
print(response)
```

```python
type(response)
```

```python
## stroutput Parser
from langchain_core.output_parsers import StrOutputParser
output_parser = StrOutputParser()
chain = prompt | llm | output_parser

response = chain.invoke({"input": "Can you tell me about langsmith?"})
print(response)
```

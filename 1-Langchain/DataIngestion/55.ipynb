```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text = "This is tutorial on OPENAI embeddings"
query_results=embeddings.embed_query(text)
print(query_results)
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
embeddings_1024
```

```python
text = "This is tutorial on OPENAI embeddings"
query_results=embeddings_1024.embed_query(text)
print(query_results)
```

```python
# using hardcode values
query_results = [random.uniform(-1.0, 1.0) for i in range(1024)]
print(len(query_results))
```

```python
## Embeded the entire document

from langchain_community.document_loaders import TextLoader

loader = TextLoader("speech.txt")
docs = loader.load()
print(docs[0].page_content)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
final_documents = text_splitter.split_documents(docs)
final_documents[0].page_content
```

```python
## Vector Embedding And Vector StoreDB

from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(final_documents, embeddings_1024)
```

```python
query = "It will be all the easier for us to conduct ourselves as belligerents"
retrived_docs = db.similarity_search(query)
print(retrived_docs)
```
